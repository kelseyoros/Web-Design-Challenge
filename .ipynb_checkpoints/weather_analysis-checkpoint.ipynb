```python
#import dependencies
import pandas as pd
import scipy.stats as st
```

```python
#Load and read data
weather_csv = "WebVisualizations/Resources/cities.csv"
weather_data = pd.read_csv(weather_csv)

weather_data.head()
```

```python
#Calculate the correlations
corr=round(st.pearsonr(weather_data['Lat'],weather_data['Cloudiness'])[0],2)
print(f"\nThe correlation between latitude and cloudiness is {corr}")

corr=round(st.pearsonr(weather_data['Lat'],weather_data['Wind Speed'])[0],2)
print(f"\nThe correlation between latitude and wind speed is {corr}")

corr=round(st.pearsonr(weather_data['Lat'],weather_data['Humidity'])[0],2)
print(f"\nThe correlation between latitude and Humidity is {corr}\n")
```
